```python
class Bank_Account():
    def __init__(self, owner, age, gender, balance):
        self.owner = owner
        self.age = age
        self.gender = gender
        self.balance = balance
        
    
    def deposit(self, deposit):
        self.balance += deposit
                
        
    def withdraw(self, withdraw):
        self.balance -= withdraw
        
        
    def check_balance(self):
        print("Your Balance {} TL".format(self.balance))
       
```

```python
owner = input("Please Enter Your Name: ")
age = int(input("Please Enter Your Age: "))
gender = input("Please Enter Your Gender: ")
balance = int(input("Please Enter Your Balance: "))
bank_account = Bank_Account(owner, age, gender, balance)

if (gender == "Female") and (21 <= age <= 30):
        print("\nWelcome To Our Bank. We Add 5.000 TL Welcome Bonus In Your Account.")
        bank_account.deposit(5000)
```

```python
print("""----------------------
    Account Transactions

    1. Deposit

    2. Withdraw

    3. Check Balance

Press "*" for terminating the operation.
----------------------""")

while True:
    
    if age < 21:
        print("Sorry You Cannot Perform This Operation.")
        break
        
        
    else:
        
        operation = input("Please Enter An Operation: ")
        
        if operation == "*":
            
            print("\nProcess Terminates... Have A Nice Day.")
            break
        
        elif operation == "1":
            deposit_amount = int(input("Please Enter Amount: "))
            
            bank_account.deposit(deposit_amount)
        
        elif operation == "2":
            withdraw_amount = int(input("Please Enter Amount: "))
            
            if withdraw_amount > bank_account.balance:
                print("Sorry You Cannot Perform This Operation.")
                continue
            
            else:
                bank_account.withdraw(withdraw_amount)
        
        elif operation == "3":
            bank_account.check_balance()
```
